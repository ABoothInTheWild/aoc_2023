```python
count = 0
is_found = False

start = "AAA"
end = "ZZZ"

while not is_found:
    for i in range(len(instructions)):
        ins = instructions[i]
        start = paths[start][ins]
        count += 1
        
        if start == end:
            is_found = True
            break
            
print(count)
```

```python
test = """LR

11A = (11B, XXX)
11B = (XXX, 11Z)
11Z = (11B, XXX)
22A = (22B, XXX)
22B = (22C, 22C)
22C = (22Z, 22Z)
22Z = (22B, 22B)
XXX = (XXX, XXX)"""
lines = test.split("\n")
lines
```

```python
paths = {}
instructions = lines[0]
starts = []

for x in lines[2:]:
    data = x.split(" = ")
    paths[data[0]] = {"L": data[1].split(", ")[0].strip("("),
              "R": data[1].split(", ")[1].strip(")")}
    
    if data[0][-1] == "A":
        starts.append(data[0])
    
paths
```

```python
steps = []

for start in starts:
    count = 0
    is_found = False
    
    while not is_found:
        for i in range(len(instructions)):
            ins = instructions[i]
            start = paths[start][ins]
            count += 1

            if start[-1] == "Z":
                is_found = True
                break
            
    steps.append(count)
```

```python
math.lcm(*steps)
```
